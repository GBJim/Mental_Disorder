```python
from pymongo import MongoClient
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def getEmptyEmotions():
    return {"joy":0,"sadness": 0,"fear":0, "anticipation": 0, "anger":0, "trust": 0, "disgust": 0 ,"surprise" : 0}

def getUsersEmotions(collectionName):
    collection = MongoClient("localhost", 27017)['idea'][collectionName]
    usersEmotions = {}
    for tweet in collection.find():
        userID = tweet["user"]["id"]
        emotion =  tweet["emotion"]["groups"][0]["name"]
        date = tweet["created_at"]
        text = tweet['text']
        if userID not in usersEmotions:
            usersEmotions[userID] = {}
        if date not in usersEmotions[userID]:
            usersEmotions[userID][date] = getEmptyEmotions()
        usersEmotions[userID][date]['text'] = text
        usersEmotions[userID][date][emotion] += 1
    return usersEmotions
        
def invalid_removal(data):
    data = np.array(data)
    return data[np.isfinite(data)]
    
def reject_outliers(data, m=2):
    data = np.array(data)
    data = data[np.isfinite(data)]
    return data[abs(data - np.mean(data)) < m * np.std(data)]        
        
        
    
        

    
```

```python



def getFlipsDuration(timeSeries):
    flips = getFlips(timeSeries)
    left_index = timeSeries.index[:-1:]
    right_index = timeSeries.index[1::]   #fixing
    return (right_index[flips].values - left_index[flips].values).astype('timedelta64[s]') / np.timedelta64(60, 's')
    
    
    


def timeSeriesTransform(usersEmotions, resample = 'W'):
    for userID in usersEmotions:
        usersEmotions[userID] = pd.DataFrame.from_dict(usersEmotions[userID], orient='index').fillna(0)
        usersEmotions[userID]['pos'] = usersEmotions[userID]['joy'] + usersEmotions[userID]['trust'] +usersEmotions[userID]['anticipation']
        usersEmotions[userID]['neg'] = usersEmotions[userID]['fear'] + usersEmotions[userID]['sadness'] + usersEmotions[userID]['disgust'] + usersEmotions[userID]['anger']
        usersEmotions[userID]['polarity'] =  usersEmotions[userID]['pos'] - usersEmotions[userID]['neg']
        
    return list(usersEmotions.values())

def polarityPlot(timeSeries, attribute = 'polarity'):
    fig, ax = plt.subplots()
    ax.set_ylabel("Sentiment Posts",  fontsize=14)
    ax.set_xlabel("Week",  fontsize=14)
    plt.plot(timeSeries[attribute])
    
    plt.axhline(y=0,linestyle="--", color="w")

    plt.show()
def peaksCount(timeSeries, attribute = 'polarity' ):
    middle = timeSeries[attribute][1:-1]
    left_elements = np.roll(timeSeries[attribute], 1)[1:-1]
    right_elements = np.roll(timeSeries[attribute], -1)[1:-1]
    local_maximum = (middle > left_elements) & (middle > right_elements)
    local_minimum = (middle < left_elements) & (middle < right_elements)
    return np.sum(local_maximum) , np.sum(local_minimum)

def crossingCount(timeSeries, attribute = 'polarity', n = 0):
    left_elements = np.roll(timeSeries[attribute], 1)[1:]
    right_elements = np.roll(timeSeries[attribute], -1)[:-1]
    positive_corssing = (left_elements < n) & (right_elements > n)
    negative_crossing = (left_elements > n) & (right_elements < n)
    return np.sum(positive_corssing | negative_crossing)


def getFlips(timeSeries):
    polarity = timeSeries['polarity'].values[:-1]
    right_elements = timeSeries['polarity'].values[1:]
    flips = (polarity * right_elements) < 0
    return flips


def getTimeRange(timeSeries):
    time_range = (np.max(timeSeries.index.values) -np.min(timeSeries.index.values) ).astype('timedelta64[s]')
    time_range = time_range / np.timedelta64(3600, 's')
    return time_range


def averageFlipsTime(timeSeries):
    time_range = (np.max(timeSeries.index.values) -np.min(timeSeries.index.values) ).astype('timedelta64[D]')
    time_range = time_range / np.timedelta64(1, 'D')
    return np.sum(getFlips(timeSeries)) / time_range

def averageFlips_posts(timeSeries):
    return np.sum(getFlips(timeSeries)) / timeSeries['polarity'].size


def get_conditional_flips(timeSeries, emotion, counter_emotion):
   
    timeSeries['polarity'] = timeSeries[emotion] - timeSeries[counter_emotion]

    duration = getFlipsDuration(timeSeries)
    
    left_index = timeSeries.index[:-1]
    right_index = timeSeries.index[1:]
    return timeSeries['polarity'], duration
    

def frequencyScatter(groups,x_name='polarity',y_name='polarity', legends = ['Normal', "BPD"], color = ['g','r']):
    for i, group in enumerate(groups):
        group_x = []
        group_y = []
        count = 0
        for timeSeries in group:
            
            #x_minimum, x_maximum = peaksCount(timeSeries, attribute=x_name)
            #y_minimum, y_maximum = peaksCount(timeSeries, attribute=y_name)
          
        
            x_density = averageFlipsTime(timeSeries)
            if x_density > 20:
                continue
            group_x.append(x_density)
            y_density = 0.5
            group_y.append(y_density)
            
            
        plt.scatter(group_x,group_y,color=color[i])
    plt.legend(legends)
    plt.xlabel(x_name, fontsize = 14)
    plt.ylabel(y_name, fontsize = 14)
    plt.show()
    print(count)
```

```python

def flipsPlot_condition_post(groups, legends = ["BPD", "Normal"], summary=True):
    
    
    counter_emotions = {"joy":'sadness',"sadness": 'joy',"fear":'anger', "anticipation": 'surprise', "anger":'fear', "trust": 'disgust', "disgust": 'trust' ,"surprise" : 'anticipation'}
    emotions = ['joy','anticipation', 'anger','trust','surprise', 'fear', 'sadness', 'disgust']
    colorTable = {"joy":"#FADB4D","sadness": "#729DC9","fear":"#35A450", "anticipation": "#F2993A", "anger":"#E43054", "trust": "#99CC33", "disgust": "#9F78BA" ,"surprise" : "#3FA5C0"}
    colors = [colorTable[emotion] for emotion in emotions]
    half_emotions = emotions[:4]
    
    if not summary:
    
        for g,group in enumerate(groups):

            all_flips = []
            for emotion in  half_emotions:    #fixing
                all_flips.append([])

            for timeSeries in group:
                for i, emotion in enumerate(half_emotions):
                    timeSeries['polarity'] = timeSeries[emotion] - timeSeries[counter_emotions[emotion]]
                    flips = getFlips(timeSeries)
                    flip_per_post = np.sum(flips) / timeSeries.shape[0]
                    all_flips[i].append(flip_per_post)


            for i, emotion in enumerate(half_emotions):
                all_flips[i] = reject_outliers(all_flips[i])


                print("{0:} to {1:} Flips duration Mean {2:.2f}, Std: {3:.2f}".format(emotion,counter_emotions[emotion],np.mean(all_flips[i]), np.std(all_flips[i])))




            plt.title(legends[g])
            plt.xlabel("Flips per post")
            #return all_flips
            plt.hist(all_flips,color=colors[0:4], edgecolor='none', bins=100)

            plt.show()
            
    else:
        for g,group in enumerate(groups):

            all_flips = []
        

            for timeSeries in group:
                for i, emotion in enumerate(half_emotions):
                    timeSeries['polarity'] = timeSeries[emotion] - timeSeries[counter_emotions[emotion]]
                    flips = getFlips(timeSeries)
                    flip_per_post = np.sum(flips) / timeSeries.shape[0]
                    all_flips.append(flip_per_post)



            all_flips = reject_outliers(all_flips)
            plt.title(legends[g])
            plt.xlabel("Flips per post")
      
            plt.hist(all_flips, edgecolor='none',bins=100)
            print("{0:} to {1:} Flips duration Mean {2:.2f}, Std: {3:.2f}".format(emotion,counter_emotions[emotion],np.mean(all_flips), np.std(all_flips)))

            plt.show()



def flipsPlot_condition_time(groups, legends = ["BPD", "Normal"], summary=True):
    
    
    counter_emotions = {"joy":'sadness',"sadness": 'joy',"fear":'anger', "anticipation": 'surprise', "anger":'fear', "trust": 'disgust', "disgust": 'trust' ,"surprise" : 'anticipation'}
    emotions = ['joy','anticipation', 'anger','trust','surprise', 'fear', 'sadness', 'disgust']
    colorTable = {"joy":"#FADB4D","sadness": "#729DC9","fear":"#35A450", "anticipation": "#F2993A", "anger":"#E43054", "trust": "#99CC33", "disgust": "#9F78BA" ,"surprise" : "#3FA5C0"}
    colors = [colorTable[emotion] for emotion in emotions]
    half_emotions = emotions[:4]
    
    if not summary:
    
        for g,group in enumerate(groups):

            all_flips = []
            for emotion in  half_emotions:    #fixing
                all_flips.append(np.array([]))

            for timeSeries in group:
                for i, emotion in enumerate(half_emotions):
                    flips, duration = get_conditional_flips(timeSeries, emotion, counter_emotions[emotion])
                    duration = duration[duration<100]
                    all_flips[i] = np.concatenate((all_flips[i], duration))


            for i, emotion in enumerate(half_emotions):

                print("{0:} to {1:} Flips duration Mean {2:.2f}, Std: {3:.2f}, Size: {4:}".format(emotion,counter_emotions[emotion],np.mean(all_flips[i]), np.std(all_flips[i]), all_flips[i].shape[0]))




            plt.title(legends[g])
            plt.xlabel("Flips per post")
            #return all_flips
            plt.hist(all_flips,color=colors[0:4], edgecolor='none', bins=100)

            plt.show()
            
    else:
        for g,group in enumerate(groups):

            all_flips = np.array([])
        

            for timeSeries in group:
                for i, emotion in enumerate(half_emotions):
                    flips, duration = get_conditional_flips(timeSeries, emotion, counter_emotions[emotion])
                    duration = duration[duration<100]
                    all_flips = np.concatenate((all_flips, duration))




            plt.title(legends[g])
            plt.xlabel("Flips per post")
            #return all_flips
            plt.hist(all_flips, edgecolor='none',bins=100)
            print("{0:} to {1:} Flips duration Mean {2:.2f}, Std: {3:.2f}, Size: {4:}".format(emotion,counter_emotions[emotion],np.mean(all_flips), np.std(all_flips),all_flips.shape[0]))

            plt.show()
        
        
       
    #plt.hist(hist_results,color=colors, bins = 100, edgecolor='none', histtype='bar')
   
    #plt.xlabel("minutes")
    #plt.show()


def flipsPlot(groups, legends = ["BPD", "Normal"], colors =['red', 'green']):
    
    hist_results = []
    for i,group in enumerate(groups):
        hist_result = []
        flips_counts = []
        tweets_counts = []
        for timeSeries in group:
            flips = np.sum(getFlips(timeSeries))
            tweet_count = timeSeries['polarity'].shape[0]
            
            if np.isfinite(tweet_count) and tweet_count > 0:
                flips_per_post = flips / tweet_count

                hist_result.append(flips_per_post)
                flips_counts.append(flips)
                tweets_counts.append(tweet_count)
            
        
        hist_result = reject_outliers(hist_result)
        hist_results.append(hist_result)
        plt.title(legends[i])
        plt.xlabel("Flips per post")
        plt.hist(hist_result,color=colors[i], bins = 100, edgecolor='none')
        plt.show()
    
        print("Tweets Mean {0:.2f} ".format(np.mean(tweets_counts)))
        print("Flips Mean {0:.2f} ".format(np.mean(flips_counts)))

        print("Flips per post Mean {0:.2f}".format(np.mean(hist_result)))
        print("Flips per post Standard Deviation {0:.2f}".format(np.std(hist_result)))
    plt.hist(hist_results,color=colors, bins = 100, edgecolor='none', histtype='bar', normed=1, stacked=True)
   
    plt.xlabel("minutes")
    plt.show()
```

```python
def foureirPlot(timeSeries):
    fig, ax = plt.subplots()
    ax.set_ylabel("Post",  fontsize=14)
    ax.set_xlabel("Time (Week as unit)",  fontsize=14)
    timeSeries['pos'].plot()
    plt.show()
    fig, ax = plt.subplots()
    ax.set_ylabel("Amplitude",  fontsize=14)
    ax.set_xlabel("Frequency (1/Week)",  fontsize=14)
    plt.plot(np.fft.rfftfreq(timeSeries['pos'].size), np.abs(np.fft.rfft(timeSeries['pos'])))
    plt.show()
```

```python
def getUsersCount(collectionName):
    collection = MongoClient("localhost", 27017)['idea'][collectionName]
    usersCount = {}
    for tweet in collection.find():
        userID = tweet['user']['id']
        usersCount[userID] = usersCount.get(userID, 0) + 1
    return usersCount
```
